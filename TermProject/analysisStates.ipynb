Jupyter notebook for analyzing the jobs for a particular zipcode. 
Dice.com provides a convinient REST API to retrieve jobs in `json` format. The jobs retrieved from Dice API are cleaned and stored in MongoDB as json documents. MongoDB's aggregation framework is used to do the analysis. Python 3.6 along with Pymongo driver is used to interact with MongoDB. Finally, plots are displayed using Matplotlib and seaborn packages.

Steps to run the analysis for a particular jobtype (eg: "Data Scientist") in a 40 mile radius of zipcode (eg: "07059")

Assumptions: 
1. Mongodb is installed and running on port default port `27017`
2. Python 3 is installed and pymongo 3.4 is used (other 3.+ versions should also work fine)
3. Matplotlib and seaborn packages are installed. (should already be installed if you are using anaconda python 3)

Step 1: Launch a new session of Jupyter Notebook or Python terminal (if using python terminal, the bar charts may not be displayed). Alternatively, you can use any IDE like pyCharm/Rodeo. 

Step 2: Import the analysis classes

```python
from analysisState import AnalysisState
```

Step 3: Create the desired analysis object by giving the jobtype and location. Next invoke retrieveAndStoreJobs to get jobs from Dice API and store them into MongoDB

```python
# Create analysis object and pass the jobtype and zip code
state = AnalysisState("Data Scientist", "07059")

# retrieve jobs from api and store them in mongodb
state.retrieveAndStoreJobs()
```

The `state.retrieveAndStoreJobs()` call, retrieved jobs from the dice API using the above mentioned URL. It then drops the existing collection, if any, to start fresh. After the jobs are successfully inserted into MongoDB's jobs database -> datascientist07059 collection, the count of the number of jobs inserted is displayed.

Step 4: Get the top 20 companies hiring a particular skill in a 40 mile radius of zipcode. Notice here, if you call `state.getTop20Companies()` without any arguments, then by default a barplot showing the top 20 companies is displayed. If you don't want the barplot to be displayed and just want the dataframe to be printed, you can pass `barplot=False`.

```python
state.getTop20Companies(barplot=False)
```
